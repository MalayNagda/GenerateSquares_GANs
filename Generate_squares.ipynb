```python
import cv2
import numpy as np
from PIL import Image
import time
import matplotlib.pyplot as plt

y=[]
#Generate random colored and random-sized squares
for i in range(500):
    #Generate a random color using random values for each- r,g,b channel 
    r=np.random.randint(255)
    g=np.random.randint(255)
    b=np.random.randint(255)
    #Generate a random-sized square 
    x=np.random.randint(45)
    img2 = np.zeros(shape=[x, x,3], dtype=np.uint8)
    img2[:]=(r, g, b)
    y.append(img2)
```

```python
fig = plt.figure()
for idx in range(8):
    ax = fig.add_subplot(2, 4, idx+1) 
    ax.imshow(y[np.random.randint(480)]) 
plt.show()
```

```python
for i in range(500):
    if not len(y[i])==0:
        cv2.imwrite("./Squares/Image number {}.jpg".format(i), y[i])
```

```python
import glob
images = glob.glob('./Squares/*.jpg')
black_sq = np.zeros(shape=[64, 64,3], dtype=np.uint8)
cv2.imwrite('./Squares/Black_Sq.jpg',black_sq)

X=[]
for image in images:
    with open(image, 'rb') as file:
        color_sq = Image.open(file)
        copy_color_sq = Image.open(file)
        copy_color_sq = np.asarray(copy_color_sq)
        num_rows, num_cols = copy_color_sq.shape[:2]
        black_sq = Image.open('./Squares/Black_Sq.jpg')
        #superimpose color square on black square
        black_sq.paste(color_sq, (32-int(num_cols/2), 32-int(num_rows/2)))
        black_sq.save('./Squares/Combined.jpg')
        combined_sq = Image.open('./Squares/Combined.jpg')
        deg=np.random.randint(180)
        #rotate the combined square by random degrees
        combined_sq=combined_sq.rotate(deg)
        combined_sq.save('./Squares/Colour_Sq.jpg')
        combined_sq = Image.open('./Squares/Colour_Sq.jpg')
        combined_sq=np.asarray(combined_sq)
        X.append(combined_sq)      
```

```python
fig = plt.figure()
for idx in range(8):
    ax = fig.add_subplot(2, 4, idx+1)
    ax.imshow(X[np.random.randint(480)]) 
plt.show()
```

```python
for i in range(len(X)):
    cv2.imwrite("./Squares/Image number {}.jpg".format(i), X[i])
```
